# **Python Basics Questions**

1. What is Python, and why is it popular?
  - Python is a high-level, interpreted, general-purpose programming language known for its simplicity and readability.

2. What is an interpreter in Python?
  - An interpreter executes Python code line-by-line, converting it into machine-readable instructions.
Python uses an interpreter (like CPython) instead of a compiler, which makes debugging easier.

3. What are pre-defined keywords in Python?
  - Keywords are reserved words in Python that have predefined meanings and cannot be used as identifiers (variable names).
Example: if, for, while, True, None, def.

4. Can keywords be used as variable names?
  - No, keywords are reserved for special purposes in the language.

5. What is mutability in Python?
  - Mutability means whether the content of an object can be changed after creation.
  - Mutable → Can be changed (e.g., lists, dictionaries)
  - Immutable → Cannot be changed (e.g., strings, tuples)

6. Why are lists mutable, but tuples are immutable?
  - Lists store elements in dynamic memory, allowing modification.
  - Tuples store elements in a fixed memory layout for faster access and safety, making them unchangeable.

7. What is difference between == and is operators in Python?
  - == → Compares values of two objects.
  - is → Compares memory addresses (checks if both variables refer to the exact same object).

8. What are logical operators in Python?
  - Logical operators are used to combine conditional statements:
  - and → True if both conditions are true
  - or → True if at least one condition is true
  - not → Reverses the logical state

9. What is type casting in Python?
  - Type casting is converting one data type into another.

10. What is difference between implicit and explicit type casting?
  - Implicit (Type Conversion) → Done automatically by Python (e.g., int + float → float)
  - Explicit (Type Casting) → Done manually using functions like int(), float(), str().

11. What is the purpose of conditional statements in Python?
  - They allow decision-making in programs by executing different code blocks depending on conditions.

12. How does the elif statement work?
  - elif means else if — used to check multiple conditions in sequence.Only the first true condition’s block runs.

13. What is the difference between for and while loops?
  - for loop → Runs a fixed number of times (good for sequences)
  - while loop → Runs until a condition becomes false (good for unknown repetitions)

14. Dexcribe a scenario where while loop is more suitable?
  - When you don’t know beforehand how many times to run the loop — e.g., reading user input until they type "exit".
